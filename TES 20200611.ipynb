```python
import plotly.express as px
import pandas as pd
```

```python
data00 = pd.read_excel('공정 시간에 따른 데이터 정리.xlsx')
```

```python
df_20200611_1_0_EPD = pd.DataFrame(data00[['index','685nm로 구한 Electrom tempreature','703nm로 구한 Electrom tempreature']])
```

```python
fig = px.line(df_20200611_1_0_EPD, x="index", y=["685nm로 구한 Electrom tempreature", "703nm로 구한 Electrom tempreature"], title="전자온도"
)
fig.update_yaxes(title_text='Electron temperature (eV)')
fig.update_xaxes(title_text='time')
fig.show()
```
